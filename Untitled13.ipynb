```python
# üì¶ Step 1: Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
# üìÇ Step 2: Define Function to Load Dataset
def load_data():
    iris = load_iris()
    X = iris.data
    y = iris.target
    feature_names = iris.feature_names
    class_names = iris.target_names
    return X, y, feature_names, class_names
```

```python
# üîÄ Step 3: Define Function to Split Dataset
def split_data(X, y, test_size=0.3, random_state=42):
    return train_test_split(X, y, test_size=test_size, random_state=random_state)
```

```python
# üß† Step 4: Define Function to Train Decision Tree
def train_decision_tree(X_train, y_train, criterion='entropy', max_depth=3):
    clf = DecisionTreeClassifier(criterion=criterion, max_depth=max_depth, random_state=42)
    clf.fit(X_train, y_train)
    return clf
```

```python
# üìä Step 5: Define Function to Evaluate Model
def evaluate_model(clf, X_test, y_test):
    y_pred = clf.predict(X_test)
    print("‚úÖ Accuracy Score:", accuracy_score(y_test, y_pred))
    print("\nüìã Classification Report:\n", classification_report(y_test, y_pred))
    print("üìâ Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# üå≥ Step 6: Define Function to Visualize Decision Tree
def visualize_tree(clf, feature_names, class_names):
    plt.figure(figsize=(12, 8))
    plot_tree(
        clf,
        feature_names=feature_names,
        class_names=class_names,
        filled=True,
        rounded=True,
        fontsize=12
    )
    plt.title("üåø Decision Tree Visualization")
    plt.show()
```

```python
# üîÅ (Optional) Step 7: Hyperparameter Tuning Function
def test_different_depths(X_train, X_test, y_train, y_test, max_depth_range):
    scores = []
    for depth in max_depth_range:
        clf = train_decision_tree(X_train, y_train, max_depth=depth)
        score = clf.score(X_test, y_test)
        scores.append(score)
        print(f"Max Depth = {depth} ‚Üí Accuracy = {score:.2f}")
    return scores
```

```python
# Run Pipeline
X, y, feature_names, class_names = load_data()
X_train, X_test, y_train, y_test = split_data(X, y)

clf = train_decision_tree(X_train, y_train)
evaluate_model(clf, X_test, y_test)
visualize_tree(clf, feature_names, class_names)

# Optional: Tune max_depth from 1 to 5
scores = test_different_depths(X_train, X_test, y_train, y_test, range(1, 6))
```
